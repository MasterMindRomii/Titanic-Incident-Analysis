```python
plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
Name = 4 + np.random.normal(0, 2, 24)
Ticket = 4 + np.random.normal(0, 2, len(Name))  # Ensure the length of Ticket matches the length of Name
sizes = np.random.uniform(15, 80, len(Name))
colors = np.random.uniform(15, 80, len(Name))

# plot
fig, ax = plt.subplots()

ax.scatter(x=Name, y=Ticket, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
